# Javascript Object Notation

## Introduction

This notebook would introduce you to Javascript Object Notation, or JSON, and how to work with them in Python. 

JSON is an agreed-upon standard way to structure data. It is meant to be readable and writeable by humans, but in practice, we seldom hand-write JSON. JSON has its roots in Javascript, and nowadays, JSON is used primarily as a way to transmit data between servers and web application. Given its human-readable characteristics, web APIs also favour its usage as a means of encoding data.

## JSON Objects

JSON is basically a set of formatting rules on how to represent data. At the most fundamental level, data is basically a set of attributes and value pairs. Take for example, a bank might have data on its customers. Customers would then have certain attributes:
>customerID

>Customer Name

>Account Balance

>Credit Rating

>Amount Owed

We can represent Customer as a JSON:
>```
{
    "customerID": "C01234",
    "Customer Name": "Average Joe",
    "Account Balance": 1234,
    "Credit Rating": "AAA",
    "Amount Owed": 456
}
```

<br>

Let's break it down a bit. The curly bracket, `{` and `}`, tells us that the data is an object. An object is nothing more than a bundle of attribute-value pairs. In this case, we have 5 pairs. And that's it. We have an object, a customer, with 5 attributes. 

Attributes are simply names given to the values, and values are the actual data. There are 2 types of data - string data and numbers 
<br><br>

## JSON Arrays

But what if we want to transmit *more* objects? 

We can create an array:

>```
[
    {
        "customerID": "C05678",
        "Customer Name": "Plain Jane",
        "Account Balance": 5678,
        "Credit Rating": "AAA",
        "Amount Owed": 789
    },
    {
        "customerID": "C05678",
        "Customer Name": "Plain Jane",
        "Account Balance": 5678,
        "Credit Rating": "AAA",
        "Amount Owed": 789
    }
]
```

<br>
Notice the square brackets, `[` and `]`. They indicate that what's in between them, is an array, or a list of objects. We have 2 objects in our case, and they are separated by a comma, `,`.
